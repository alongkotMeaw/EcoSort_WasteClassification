```python

# run_on_gogle_colap

!pip install bing-image-downloader --quiet

from bing_image_downloader import downloader
import os
from pathlib import Path
import shutil


keywords = {
    "straw": [
        "plastic drinking straw white background",
        "colorful straws isolated",
        "paper straw eco friendly",
        "used plastic straw waste",
        "straws on beach pollution",
        "bent straw close up",
        "straws in cup drink",
        "metal reusable straw",
        "plastic straw garbage",
        "biodegradable straw"
    ]
}


#  โหลดภาพจาก Bing
for label, kw_list in keywords.items():
    save_dir = f"dataset_real/{label}"
    os.makedirs(save_dir, exist_ok=True)
    print(f" เริ่มโหลดภาพ: {label}")
    for i, kw in enumerate(kw_list):
        print(f"   keyword {i+1}: {kw}")
        downloader.download(
            keyword=kw,
            limit=100,
            output_dir=save_dir,
            adult_filter_off=True,
            force_replace=False,
            timeout=60
        )
    print(f" เสร็จสิ้น: {label}\n")

#  รวมไฟล์จาก subfolder มาไว้ที่เดียว
for label in keywords:
    class_dir = Path(f"dataset_real/{label}")
    for subfolder in class_dir.iterdir():
        if subfolder.is_dir():
            for file in subfolder.glob("*"):
                shutil.move(str(file), str(class_dir))
            os.rmdir(subfolder)

print(" รูปทั้งหมดถูกดาวน์โหลดและรวมไว้ใน dataset_real/ พร้อมใช้งานแล้ว")
```
